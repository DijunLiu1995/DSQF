In order to make it easier to merge the election dataset with other datasets, we added FIPS and state abbreviation to the original dataset scrapped from the web pages.

```python
import pandas as pd
```

This is a state-abbreviation mapping downloaded online.

```python
state_abbr = pd.read_csv('state_abbr.csv')
state_abbr.head()
```

Here we used the data given in assignment 1 to create FIPS for each county.

```python
data = pd.read_csv('national_county.txt', sep=",", header=None, dtype=object)
data.columns = ["STATE", "STATEFP", "COUNTYFP", "COUNTYNAME","CLASSFP"]
data['FIPS'] = pd.to_numeric(data['STATEFP']+data['COUNTYFP'],\
                             downcast='integer')
data.head()
```

```python
data['COUNTY'] = [' '.join(i[0:-1]) for i in data['COUNTYNAME'].str.split(' ')]
data.head()
```

```python
def to_FIPS(path,name):
    Election = pd.read_csv(path)
    Election['state_up'] = Election['state'].str.replace('-', ' ').str.upper()
    Election = Election.merge(state_abbr, left_on='state_up', right_on='State', how='left')
    Election = Election.merge(data, left_on=['Abbreviation','county'], right_on=['STATE','COUNTY'], how='left')
    Election = Election.drop(['state_up','State','STATE','STATEFP','COUNTYFP','COUNTYNAME','CLASSFP','COUNTY'], axis=1)
    Election.to_csv(name)
```

```python
to_FIPS('elections/2000/general.csv', 'general_FIPS.csv')
```
