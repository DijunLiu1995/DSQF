```python
import pandas as pd
import string
```

```python
def reference_replacer(df):
#     df=df.copy()
    for c in df.columns:
        print(c)
        
        if df[df[c].astype('str').str.contains("\[")==True].shape[0]>0:

            for letter_ in list(string.ascii_lowercase):
                df[c] = df[c].astype('str').str.replace("\[" + letter_ + "\]",'')


            for num in range(0,100):
                df[c] = df[c].astype('str').str.replace("\[" + str(num) + "\]",'')
            
    return df
        
        
    
```

```python
p = '../geo/elections/2016/'
```

```python
save_dir='../primary_data/'
```

```python
primary_data = pd.read_csv(p + 'primary.csv')
```

```python
dem_data = pd.read_csv('../primary_data/cleaned_dems.csv')
```

```python
rep_data = pd.read_csv('../primary_data/cleaned_reps.csv')
```

```python
primary_data[(primary_data['state']=='Nebraska') & (primary_data.party1=='D')]
```

```python
dem_data = reference_replacer(dem_data.copy())
```

```python
rep_data = reference_replacer(rep_data.copy())
```

```python
cleaned_dem_state_date = dem_data.groupby('State/territory').agg({'Date':'max'}).sort_values('Date').reset_index()
dem_data = pd.merge(dem_data,cleaned_dem_state_date,on=['State/territory','Date'])
```

```python
cleaned_rep_state_date = rep_data.groupby('State').agg({'Date':'max'}).sort_values('Date').reset_index()
rep_data = pd.merge(rep_data,cleaned_rep_state_date,on=['State','Date'])
```

```python
dem_data = dem_data.rename(columns={'State/territory':'state','Date':'primary_date','Type':'primary_type'})
```

```python
rep_data = rep_data.rename(columns={'State':'state','Allocation-Contest':'primary_type','Date':'primary_date'})
```

```python
primary_data_dems = pd.merge(primary_data[primary_data.party1=='D'],dem_data[['state','primary_date','primary_type']],on='state',how='outer')
```

```python
primary_data_reps = pd.merge(primary_data[primary_data.party1=='R'],dem_data[['state','primary_date','primary_type']],on='state',how='outer')
```

```python
appended_two_party_primary_data = primary_data_dems.append(primary_data_reps)
```

```python
appended_two_party_primary_data.to_csv(save_dir + 'primary_processed.csv')
```
