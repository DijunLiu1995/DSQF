```python
def read_file(fname):
#     fname = dir_+'earnings_data/actuals.txt'
    with open(fname,'r') as f:
        data = f.readlines()
        
    regex = re.compile(r"\s+", re.IGNORECASE)

    out = []

    for l in data:

        l = l.strip()
        replaced = re.sub('\s+', '\s', l)

        l = (replaced.split('\s'))


        if len(l)>1:
            out.append(l) 
            
    return out
    
```
